```python
import numpy as np

import skimage.io as io
from skimage.exposure import histogram, match_histograms
import matplotlib.pyplot as plt;
```

```python
img_FS144 = io.imread('../data/unet/family_machine/FS144/train/img/GD_04092019AlgCo_Cas_Pos57_TXRED_Exp001_frame100_cut.png')
img_ZF270g = io.imread('../data/unet/family_machine/ZF270g/train/img/pos33-eGFP_dual-A10200_cut.png')
img_phase = io.imread('../data/unet/family_machine/Phase/train/img/CWRt00100xy001c1.tif', as_gray=True)

def scale_pixel_vals(img):
    '''
    Scale pivel values between 0 and 1.
    '''

    img = np.array(img)
    return ((img - img.min()) / (img.max() - img.min()))
```

```python
img_FS144 = scale_pixel_vals(img_FS144)
img_ZF270g = scale_pixel_vals(img_ZF270g)
img_phase = scale_pixel_vals(img_phase)
```

```python
plt.figure(figsize=(10,10))
img_hist, bins = histogram(img_FS144)
plt.subplot(1,2,1)
plt.imshow(img_FS144)
plt.subplot(1,2,2)
plt.plot(bins, img_hist)
```

```python
plt.figure(figsize=(10,10))
img_hist, bins = histogram(img_ZF270g)
plt.subplot(1,2,1)
plt.imshow(img_ZF270g)
plt.subplot(1,2,2)
plt.plot(bins, img_hist)
```

```python
img_phase_inv = 1 - img_phase
```

```python
plt.figure(figsize=(10,10))
img_hist, bins = histogram(img_phase_inv)
plt.subplot(1,2,1)
plt.imshow(img_phase_inv)
plt.subplot(1,2,2)
plt.plot(bins, img_hist)
```

```python
img_match = match_histograms(img_FS144, img_ZF270g)
```

```python
plt.figure(figsize=(10,10))
img_hist, bins = histogram(img_match)
plt.subplot(1,2,1)
plt.imshow(img_match)
plt.subplot(1,2,2)
plt.plot(bins, img_hist)
```
