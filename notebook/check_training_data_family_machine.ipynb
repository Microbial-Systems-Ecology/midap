This notebook has the purpose to check the training data. It is important that in the raw image the cells have a high value and the background a low value (0).

```python
import os
from skimage import io

import matplotlib.pyplot as plt

import numpy as np
```

```python
path = '/cluster/work/sis/cdss/oschmanf/ackermann-bacteria-segmentation/data/unet/family_machine/'
cell_types = os.listdir(path)
print(cell_types)
```

```python
def normalize_image(img):
    return (img - img.min())/(img.max() - img.min())

def invert_img(img):
    img = img.astype(float)
    return -1*(img - img.max())
```

# CB15-WT

```python
img_name = path + 'CB15-WT/train/img/Caulobacter_WT-Raw.tif'
seg_name = path + 'CB15-WT/train/mask/Caulobacter_WT_segmented.tif'
```

```python
img = io.imread(img_name)
seg = io.imread(seg_name)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))
im = ax[0].imshow(img)
plt.colorbar(im, ax=ax[0])

s = ax[1].imshow(seg)
plt.colorbar(s, ax=ax[1])
```
