```python
def gen_subplot(timeframe):
    cax = ax.imshow(added_labels[timeframe], cmap=plt.cm.get_cmap('RdBu', 7))
    cbar = fig.colorbar(cax, ticks=np.arange(0.5, 8, 1))
    cbar.ax.set_yticklabels(['BG', 'PH', 'GFP', 'PH+GFP', 'mCherry', 'PH+mCherry', 'GFP+mCherry', 'PH+GFP+mCherry'])
    ax.set_xticks([])
    ax.set_yticks([])
    plt.title('Timeframe ' + str(timeframe))
```
