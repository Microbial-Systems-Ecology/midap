```python
import skimage.io as io
import matplotlib.pyplot as plt
import os
from sklearn.metrics import log_loss, roc_auc_score, accuracy_score, confusion_matrix, f1_score, balanced_accuracy_score
from skimage import io
import numpy as np
import pandas as pd
#from keras import metrics
```

```python
def compute_loss(files, files_il, seg_dir, seg_il_dir):
    score = []
    for seg, seg_il in zip(files, files_il):
        y_pred = io.imread(seg_dir + seg)
        y_pred = (y_pred > 0).astype(int)
        y_true = io.imread(seg_il_dir + seg_il)
        y_true[y_true == 255] = 1
        #y_true = (y_true - 1).astype(int)
        #loss.append(log_loss(y_pred, y_true))
#         acc = roc_auc_score(y_true.flatten(), y_pred.flatten())
        acc = f1_score(y_true.flatten(), y_pred.flatten())
        #acc = metrics.binary_accuracy(y_true, y_pred)
        #print(confusion_matrix(y_true.flatten(), y_pred.flatten()))
        #print(np.unique(y_true.flatten()), np.unique(y_pred.flatten()))
        score.append(acc)
    return score
```

```python
#path = '../data/data_Glen/R2-Mumbai/5_13B01g-FS144r/pos25/'
path = '/cluster/work/sis/oschmanf/R2-Mumbai/5_13B01g-FS144r/pos25/'

def load_cut(pos, frame, ch):
    return io.imread(path + '/cut/'+ch+'/'+pos+'-'+ch+'_dual-A10'+frame+'_cut.png')

def load_seg(pos, frame, ch):
    return io.imread(path + '/seg_unet/'+ch+'/png/'+pos+'-'+ch+'_dual-A10'+frame+'_seg.png')
```

```python
def gen_fig(pos, ch):
    frame1 ='000'
    frame2 ='100'
    frame3 ='200'
    
    plt.figure(figsize = (15,7))
    plt.subplots_adjust(wspace = 0.1, hspace = 0.05)
    plt.subplot(2,3,1)
    plt.imshow(load_cut(pos, frame1, ch), cmap = 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame1))

    plt.subplot(2,3,4)
    plt.imshow(load_seg(pos, frame1, ch) > 0)
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame1))

    plt.subplot(2,3,2)
    plt.imshow(load_cut(pos, frame2, ch), cmap = 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame2))

    plt.subplot(2,3,5)
    plt.imshow(load_seg(pos, frame2, ch) > 0)
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame2))

    plt.subplot(2,3,3)
    plt.imshow(load_cut(pos, frame3, ch), cmap = 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame3))

    plt.subplot(2,3,6)
    plt.imshow(load_seg(pos, frame3, ch) > 0)
    plt.xticks([])
    plt.yticks([])
    plt.title('frame ' + str(frame3))
    plt.suptitle(ch);
```

```python
#path = '../data/data_Glen/R2-Mumbai/5_13B01g-FS144r/pos25/'
path = '/cluster/work/sis/oschmanf/R2-Mumbai/5_13B01g-FS144r_results/pos25/'
seg_dir_ph = path + 'seg/PH/png/'
seg_dir_eGFP = path + 'seg/eGFP/png/'
seg_dir_mCherry = path + 'seg/mCherry/png/'

seg_edge_ph = path + 'seg_edge/PH/png/'
seg_edge_eGFP = path + 'seg_edge/eGFP/png/'
seg_edge_mCherry = path + 'seg_edge/mCherry/png/'

seg_region_ph = path + 'seg_region_based/PH/png/'
seg_region_eGFP = path + 'seg_region_based/eGFP/png/'
seg_region_mCherry = path + 'seg_region_based/mCherry/png/'

seg_kmeans_region_ph = path + 'seg_kmeans_region_based/PH/png/'
seg_kmeans_region_eGFP = path + 'seg_kmeans_region_based/eGFP/png/'
seg_kmeans_region_mCherry = path + 'seg_kmeans_region_based/mCherry/png/'

seg_enh_cont_ph = path + 'seg_enh_cont/PH/png/'
seg_enh_cont_eGFP = path + 'seg_enh_cont/eGFP/png/'
seg_enh_cont_mCherry = path + 'seg_enh_cont/mCherry/png/'

#seg_enh_cont_ph = '../data/data_Glen/R2-Mumbai/5_13B01g-FS144r/pos25/seg_enh_cont/PH/png/'
seg_unet_eGFP = path + '/eGFP/seg_unet2/'
seg_unet_mCherry = path + '/mCherry/png/'

seg_il_dir_ph = path + 'seg_ilastik/PH/'
seg_il_dir_eGFP = path + 'seg_ilastik/eGFP/'
seg_il_dir_mCherry = path + 'seg_ilastik/mCherry/'
```

```python
files_ph = np.sort(os.listdir(seg_dir_ph))
files_eGFP = np.sort(os.listdir(seg_dir_eGFP))
files_mCherry = np.sort(os.listdir(seg_dir_mCherry))

files_edge_ph = np.sort(os.listdir(seg_edge_ph))
files_edge_eGFP = np.sort(os.listdir(seg_edge_eGFP))
files_edge_mCherry = np.sort(os.listdir(seg_edge_mCherry))

files_region_ph = np.sort(os.listdir(seg_region_ph))
files_region_eGFP = np.sort(os.listdir(seg_region_eGFP))
files_region_mCherry = np.sort(os.listdir(seg_region_mCherry))

files_kmeans_region_ph = np.sort(os.listdir(seg_kmeans_region_ph))
files_kmeans_region_eGFP = np.sort(os.listdir(seg_kmeans_region_eGFP))
files_kmeans_region_mCherry = np.sort(os.listdir(seg_kmeans_region_mCherry))

files_enh_cont_ph = np.sort(os.listdir(seg_enh_cont_ph))
files_enh_cont_eGFP = np.sort(os.listdir(seg_enh_cont_eGFP))
files_enh_cont_mCherry = np.sort(os.listdir(seg_enh_cont_mCherry))

files_unet_eGFP = np.sort(os.listdir(seg_unet_eGFP))
files_unet_mCherry = np.sort(os.listdir(seg_unet_mCherry))

files_il_ph = np.sort(os.listdir(seg_il_dir_ph))
files_il_eGFP = np.sort(os.listdir(seg_il_dir_eGFP))
files_il_mCherry = np.sort(os.listdir(seg_il_dir_mCherry))
```

```python
#loss for combined segmentation
loss_ph = compute_loss(files_ph, files_il_ph, seg_dir_ph, seg_il_dir_ph)
mean_loss_ph = np.mean(loss_ph)

loss_eGFP = compute_loss(files_eGFP, files_il_eGFP, seg_dir_eGFP, seg_il_dir_eGFP)
mean_loss_eGFP = np.mean(loss_eGFP)

loss_mCherry = compute_loss(files_mCherry, files_il_mCherry, seg_dir_mCherry, seg_il_dir_mCherry)
mean_loss_mCherry = np.mean(loss_mCherry)
```

```python
# loss for edge segmentation
loss_edge_ph = compute_loss(files_edge_ph, files_il_ph, seg_edge_ph, seg_il_dir_ph)
mean_loss_edge_ph = np.mean(loss_edge_ph)

loss_edge_eGFP = compute_loss(files_edge_eGFP, files_il_eGFP, seg_edge_eGFP, seg_il_dir_eGFP)
mean_loss_edge_eGFP = np.mean(loss_edge_eGFP)

loss_edge_mCherry = compute_loss(files_edge_mCherry, files_il_mCherry, seg_edge_mCherry, seg_il_dir_mCherry)
mean_loss_edge_mCherry = np.mean(loss_edge_mCherry)
```

```python
# loss for region based segmentation
loss_region_ph = compute_loss(files_region_ph, files_il_ph, seg_region_ph, seg_il_dir_ph)
mean_loss_region_ph = np.mean(loss_region_ph)

loss_region_eGFP = compute_loss(files_region_eGFP, files_il_eGFP, seg_region_eGFP, seg_il_dir_eGFP)
mean_loss_region_eGFP = np.mean(loss_region_eGFP)

loss_region_mCherry = compute_loss(files_region_mCherry, files_il_mCherry, seg_region_mCherry, seg_il_dir_mCherry)
mean_loss_region_mCherry = np.mean(loss_region_mCherry)
```

```python
# loss for kmeans + region based segmentation
loss_kmeans_region_ph = compute_loss(files_kmeans_region_ph, files_il_ph, seg_kmeans_region_ph, seg_il_dir_ph)
mean_loss_kmeans_region_ph = np.mean(loss_kmeans_region_ph)

loss_kmeans_region_eGFP = compute_loss(files_kmeans_region_eGFP, files_il_eGFP, seg_kmeans_region_eGFP, seg_il_dir_eGFP)
mean_loss_kmeans_region_eGFP = np.mean(loss_kmeans_region_eGFP)

loss_kmeans_region_mCherry = compute_loss(files_kmeans_region_mCherry, files_il_mCherry, seg_kmeans_region_mCherry, seg_il_dir_mCherry)
mean_loss_kmeans_region_mCherry = np.mean(loss_kmeans_region_mCherry)
```

```python
# loss for enhanced contrast
loss_enh_cont_ph = compute_loss(files_enh_cont_ph, files_il_ph, seg_enh_cont_ph, seg_il_dir_ph)
mean_loss_enh_cont_ph = np.mean(loss_enh_cont_ph)

loss_enh_cont_eGFP = compute_loss(files_enh_cont_eGFP, files_il_eGFP, seg_enh_cont_eGFP, seg_il_dir_eGFP)
mean_loss_enh_cont_eGFP = np.mean(loss_enh_cont_eGFP)

loss_enh_cont_mCherry = compute_loss(files_enh_cont_mCherry, files_il_mCherry, seg_enh_cont_mCherry, seg_il_dir_mCherry)
mean_loss_enh_cont_mCherry = np.mean(loss_enh_cont_mCherry)
```

```python
# loss for unet

loss_unet_eGFP = compute_loss(files_unet_eGFP, files_il_eGFP, seg_unet_eGFP, seg_il_dir_eGFP)
mean_unet_eGFP = np.mean(loss_unet_eGFP)

loss_unet_mCherry = compute_loss(files_unet_mCherry, files_il_mCherry, seg_unet_mCherry, seg_il_dir_mCherry)
mean_loss_unet_mCherry = np.mean(loss_unet_mCherry)
```

```python
df = pd.DataFrame(columns = ['PH', 'eGFP', 'mCherry'], index = ['edge detection (canny) + k-means clustering',
                                                                'edge detection (canny)', 'region-based approach',
                                                                'kmeans plus region-based approach', 
                                                                'enhanced contrast', 'U-Net'])
df.iloc[0]['PH'] = mean_loss_ph
df.iloc[0]['eGFP'] = mean_loss_eGFP
df.iloc[0]['mCherry'] = mean_loss_mCherry

df.iloc[1]['PH'] = mean_loss_edge_ph
df.iloc[1]['eGFP'] = mean_loss_edge_eGFP
df.iloc[1]['mCherry'] = mean_loss_edge_mCherry

df.iloc[2]['PH'] = mean_loss_region_ph
df.iloc[2]['eGFP'] = mean_loss_region_eGFP
df.iloc[2]['mCherry'] = mean_loss_region_mCherry

df.iloc[3]['PH'] = mean_loss_kmeans_region_ph
df.iloc[3]['eGFP'] = mean_loss_kmeans_region_eGFP
df.iloc[3]['mCherry'] = mean_loss_kmeans_region_mCherry

df.iloc[4]['PH'] = np.nan
df.iloc[4]['eGFP'] = mean_loss_enh_cont_eGFP
df.iloc[4]['mCherry'] = mean_loss_enh_cont_mCherry

df.iloc[5]['PH'] = np.nan
df.iloc[5]['eGFP'] = mean_unet_eGFP
df.iloc[5]['mCherry'] = mean_loss_unet_mCherry
```
